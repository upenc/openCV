### Arithmetic Operations

```python
image1 = cv2.imread("D:/openCV-project/openCV/Image/Kedarnath.jpg")
image2 = cv2.imread("D:/openCV-project/openCV/Image/adiyogi2.jpg")

# checking for size 
if image1.shape != image2.shape:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

added_image = cv2.add(image1, image2)
cv2.imshow("added Image",added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Blending the image

combining two images to produce a third image that shows a mixture of the two.
* BlendedImage=Image1+(1)Image2
* where 
    * 0<<1 

```python
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

alpha = 0.5

# Blend the images using addWeighted
blended_image = cv2.addWeighted(image1_rgb, alpha, image2_rgb, 1 - alpha, 0)

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

axs[0].imshow(image1_rgb)
axs[0].set_title('Image 1')
axs[0].axis('off')

axs[1].imshow(image2_rgb)
axs[1].set_title('Image 2')
axs[1].axis('off')

# Display the blended image
axs[2].imshow(blended_image)
axs[2].set_title(f'Blended Image (with alpha={alpha})')
axs[2].axis('off')

cv2.imwrite('D:/openCV-project/openCV/Image/blended_image.jpg', blended_image)
plt.tight_layout()
plt.show()
```

#### Bitwise Operations

```python
imag = np.zeros((300, 300, 3), dtype=np.uint8)
cv2.rectangle(imag, (50, 50), (250, 250), (255, 255, 255), -1)
mask = np.zeros((300, 300), dtype=np.uint8)
cv2.circle(mask, (150, 150), 100, 255, -1)

bitwise_and = cv2.bitwise_and(imag, imag, mask=mask)
bitwise_or = cv2.bitwise_or(imag, imag, mask=mask)
bitwise_xor = cv2.bitwise_xor(imag, imag, mask=mask)
bitwise_not = cv2.bitwise_not(mask)

cv2.imshow('Imag', image)
cv2.imshow('Mask', mask)
cv2.imshow('Bitwise AND', bitwise_and)
cv2.imshow('Bitwise OR', bitwise_or)
cv2.imshow('Bitwise XOR', bitwise_xor)
cv2.imshow('Bitwise NOT of Mask', bitwise_not)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Blurring

```python
img = cv2.imread('D:/openCV-project/openCV/Image/kedarnath.jpg')

averaging_blur = cv2.blur(img,(5,5))
cv2.imshow('Averaging Blur', averaging_blur)

gaussian_blur = cv2.GaussianBlur(img, (5,5),10)
cv2.imshow('Gaussian Blur', gaussian_blur)

median_blur = cv2.medianBlur(img,5)
cv2.imshow('Median Blur', median_blur)

bilateral_blur = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imshow('Bilateral Filter', bilateral_blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
