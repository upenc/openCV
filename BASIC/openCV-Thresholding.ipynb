```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

Thresholding is a technique used in image processing and computer vision to create binary images from grayscale images. It involves converting each pixel in an image to either black or white based on a certain threshold value. 

### Simple Thresholding

```python
img = cv2.imread("adiyogi2.jpg",cv2.IMREAD_GRAYSCALE)


ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
 
for i in range(6):
 plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
 plt.title(titles[i])
 plt.xticks([]),plt.yticks([])
 
plt.show()
```

In this we used one global value as a threshold. But this might not be good in all cases, e.g. if an image has different lighting conditions in different areas.

* In that case, adaptive thresholding can help.

### Adaptive Thresholding

```python
#threshold value is the mean of the neighborhood area minus the constant C.
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,13,2)

#Threshold value is the weighted sum of neighbourhood values where weights are a gaussian window.
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,13,2)

fig, axs = plt.subplots(1, 2, figsize=(10, 8))
axs[0].imshow(th2,'gray')
axs[0].set_title('ADAPTIVE_THRESH_MEAN_C')
axs[1].imshow(th3,'gray')
axs[1].set_title('ADAPTIVE_THRESH_GAUSSIAN_C')
plt.show()
```
